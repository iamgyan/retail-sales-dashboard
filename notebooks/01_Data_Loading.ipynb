```python
# Convert dates to datetime
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
df.info()
```

```python
# Create new date features

df["Order Year"] = df["Order Date"].dt.year

df["Order Month"] = df["Order Date"].dt.month_name()

df["Order Quarter"] = df["Order Date"].dt.quarter

df["Order Day"] = df["Order Date"].dt.day_name()

df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days

df.info()
```

What does create_engine() do?

It creates an Engine object, which:

Knows how to connect to your database.
Manages database connections efficiently using connection pooling.
Allows libraries like Pandas to communicate with the database.
Does not actually connect immediately. The connection is established only when you execute a query (this is called lazy connection).

```python
from sqlalchemy import create_engine
from sqlalchemy.engine import URL

url = URL.create(
    drivername="postgresql+psycopg2",
    username="postgres",
    password="sql@2025",     
    host="localhost",
    port=5432,
    database="retail_project"
)

engine = create_engine(url)

# engine = create_engine(
#     "postgresql+psycopg2://postgres:sql%402025@localhost:5432/retail_project"
# )
```
